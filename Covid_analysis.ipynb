```python
import pandas as pd
import numpy as np
import duckdb
```

```python
inp0 = pd.read_csv("/content/CovidDeaths.csv")
inp1 = pd.read_csv("/content/CovidVaccinations.csv")
```

We are going to use following data in almost many of following queries

```python
# Simple select query
query = """
Select Location, date, total_cases, new_cases, total_deaths, population
From inp0
Where continent is not null
order by 1,2
"""
duckdb.query(query).df()
```

Below is the Data for death percentage in India between available time period

```python
# Simple select query
query = """
Select Location, date, total_cases, total_deaths, (total_deaths/total_cases)*100 as DeathPercentage
From inp0
Where location like '%India%'
And continent is not null
order by 2
"""
duckdb.query(query).df()
```

Following query shows percentage of population infected with covid

```python
# Simple select query
query = """
Select Location, date, Population, total_cases,  (total_cases/population)*100 as PercentPopulationInfected
From inp0
Where location like '%India%'
order by 1,2
"""
duckdb.query(query).df()
```

following query will give countries sorted by highest percentage population infected

```python
# Simple select query
query = """
Select Location, Population, MAX(total_cases) as HighestInfectionCount,  Max((total_cases/population))*100 as PercentPopulationInfected
From inp0
Group by Location, Population
order by PercentPopulationInfected desc
"""
duckdb.query(query).df()
```

Following query will give maximum death count for each country sorted from highest to lowest

```python
# Simple select query
query = """
Select Location, MAX(cast(Total_deaths as int)) as TotalDeathCount
From inp0
Where continent is not null
Group by Location
order by TotalDeathCount desc
"""
duckdb.query(query).df()
```

Following query shows continents with highest death count sorted from highest to lowest

```python
# Simple select query
query = """
Select continent, MAX(cast(Total_deaths as int)) as TotalDeathCount
From inp0
Where continent is not null
Group by continent
order by TotalDeathCount desc
"""
duckdb.query(query).df()
```

Following query gives global total cases and total deaths with percentage of how many patients died

```python
# Simple select query
query = """
Select SUM(new_cases) as total_cases, SUM(cast(new_deaths as int)) as total_deaths, SUM(cast(new_deaths as int))/SUM(New_Cases)*100 as DeathPercentage
From inp0
where continent is not null
order by 1,2
"""
duckdb.query(query).df()
```

Following query shows Percentage of Population that has recieved at least one Covid Vaccine

```python
# Simple select query
query = """
Select inp0.continent, inp0.location, inp0.date, inp0.population, inp1.new_vaccinations
, SUM(CAST(inp1.new_vaccinations AS INTEGER)) OVER (Partition by inp0.Location Order by inp0.location, inp0.Date) as RollingPeopleVaccinated
From inp0
Join inp1
	On inp0.location = inp1.location
	and inp0.date = inp1.date
where inp0.continent is not null
order by 2,3
"""
duckdb.query(query).df()
```

Using CTE to perform Calculation on Partition By in previous query

```python
# Simple select query
query = """
With PopvsVac (Continent, Location, Date, Population, New_Vaccinations, RollingPeopleVaccinated)
as
(
Select inp0.continent, inp0.location, inp0.date, inp0.population, inp1.new_vaccinations
, SUM(CAST(inp0.new_vaccinations AS INTEGER)) OVER (Partition by inp0.Location Order by inp0.location, inp0.Date) as RollingPeopleVaccinated
--, (RollingPeopleVaccinated/population)*100
From inp0
Join inp1
	On inp0.location = inp1.location
	and inp0.date = inp1.date
where inp0.continent is not null
--order by 2,3
)
Select *, (RollingPeopleVaccinated/Population)*100
From PopvsVac
"""
duckdb.query(query).df()
```
